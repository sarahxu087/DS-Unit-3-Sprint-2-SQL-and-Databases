```python
import sqlite3
conn=sqlite3.connect('rpg_db.sqlite3')
curs = conn.cursor()
query = """SELECT *
        FROM charactercreator_character;"""
curs.execute(query)
totalCharacter = curs.fetchall()
print(totalCharacter)
```

```python
type(totalCharacter)
```

```python
rpg_character_docs = []
    


for i in range(len(totalCharacter)):        
        
    rpg_doc = {
               
    
                'sql_key':totalCharacter[i][0],
                'name':totalCharacter[i][1],
                'level':totalCharacter[i][2],
                'exp':totalCharacter[i][3],
                'hp':totalCharacter[i][4],
                'strength':totalCharacter[i][5],
                'intelligence':totalCharacter[i][6],
                'dexterity':totalCharacter[i][7],
                'wisdom':totalCharacter[i][8],
                    
    
    }

    rpg_character_docs.append(rpg_doc)
```
